```python
#import Libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots we will create 



# Read in the data 

df = pd.read_csv(r'/Users/nieveenalia/Downloads/all_folders/csv_files/movies.csv')
```

```python
df = pd.read_csv(r'/Users/nieveenalia/Downloads/all_folders/csv_files/movies.csv', header=1)
```

```python
# Lets look at the date 

for col in df.columns:
    pct_missing = np.mean(df[col].isnull()) * 100
    print(f"{col} - {pct_missing:.1f}%")

df
```

```python
# Data types for our columns 

df.dtypes
```

```python
# changing the data type of columns 

df['budget'] = df['budget'].fillna(0).astype('int64')
df['gross'] = df['gross'].fillna(0).astype('int64')

df
```

```python
cols_to_drop = ['yearcorrect', 'release_date_clean', 'release_year', 'released_clean']
df.drop(columns=[col for col in cols_to_drop if col in df.columns], inplace=True)

```

```python
df['Month'] = df['released'] .astype(str).str[:4]

df
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Drop any duplicates 
df.drop_duplicates()
```

```python
# Scatter plot with budget vs gross 

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earning')
plt.xlabel('Budget Earnings')
plt.ylabel('Gross for Film')
plt.show()
```

```python
#Plot budget vs gross using seaborn 

sns.regplot(x='budget' , y='gross', data=df,scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```
